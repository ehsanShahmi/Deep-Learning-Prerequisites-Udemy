# Line chart

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.rcParams["figure.figsize"] = "12, 8"
```

```python
x = np.linspace(0, 10, 10)
# A way to produce dataframe for x-axis...startng point, ending pont, how many points in between
y = np.sin(x)
# element by elemnt operation....so it takes every value of x and applies sine
```

```python
plt.plot(x,y)
# Here we plot the two axes x and y with the help of a line.
# Plot() function is responsible to draw lines.
```

```python
# We can also add labels for the axes and title for the graph
plt.xlabel("Time")
plt.ylabel("Some function")
plt.title("My cool chart")
# Without show function, the plot can't be viewed.
plt.show()
```

One imp fact. The curve doesn't appeear to b a sine curve...bcoz, there are less points & also all points in our computer system are discrete, as its impossible to show continuous points.

So we try with more number of points.

```python
x = np.linspace(0, 10, 100)    #100 points in between 0 and 10...whereas b4 it was 10 points in between
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

# Scatterplot

Previously, we knew the exact function for x which is sine of x. But in ML, we ususally dont know anything. There comes the scatter plot for visualization for the datapoints rather than simply known function plot.

```python
A = pd.read_csv("data_1d.csv", header=None).as_matrix()
A
```

```python
# In A, x- and y-axes are first and second columns respectively:
# We used slicing, o pick out entire columns
x = A[:,0]    
y = A[:,1]
```

```python
# Now we do the plotting
plt.scatter(x, y)
```

```python
plt.show()
```

Now we plot with a function so to see that the dataframe points actually fits.

```python
x_line = np.linspace(0 , 100, 100)
y_line = 2 * x_line + 1
plt.scatter(x, y)
plt.plot(x_line, y_line)
plt.show()
```
