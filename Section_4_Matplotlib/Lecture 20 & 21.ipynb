# Histogram

### Histogram is a discritized probability distribution of the database...done in terms of frequency counts in each bucket of values

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = "12, 8"
```

```python
A = pd.read_csv("data_1d.csv", header=None).as_matrix()
x = A[:,0]    
y = A[:,1]
```

```python
# Histogram of the x-datapoints
plt.hist(x)
```

```python
# The histogram is now showed:
plt.show()
```

### Random function gives us uniformly distributed points. We can review that using histogram

```python
R = np.random.random(10000)
plt.hist(R)
plt.show()
```

We can see here, that heights of the bars are similar, i.e the datapoints are uniformly distributed

```python
# We can modify the number of buckets(or bars) in the method call...defualt value is 10.
plt.hist(R, bins=20)
plt.show()
```

### Error of a fitted line is normally distributed, we assume always. If we subtract the value of y, from its corresponding fitted line, then the result comes as normally distributed. This can be verified using histogram

```python
y_actual = 2 * x + 1
residuals = y - y_actual
plt.hist(residuals)
plt.show()
```

Here, we subtracted the y values of actual best-fit line of that dataframe from the y values of the dataframe, which is A. The resulting residuals os taken to draw the histogram. This is assumed to become a normal distribution. This is thus also proved using histogram.

# Plotting images

```python
df = pd.read_csv("train.csv")    #no header taken
df.shape
```

Original screen is 28x28...so there are 784 pixels...each having 1column in the dataframe and a label column. And there are 42000 samples inside the dataframe.

```python
M = df.as_matrix()
# We need to find an image...so we convert it to a matrix first, so that we can access easily
```

```python
im = M[0, 1:]
# This implies zeroth row and all columns staring from 1th column
im.shape
```

This shows a vector of size 784. But we really want a matrix of 28x28, as we want an image.

```python
# Reshape() function of numpy can reshape the vector
im = im.reshape(28, 28)
im.shape
```

Thus we get our required matrix, which is now an image

```python
# now we can display this image 
plt.imshow(im)
plt.show()
```

```python
# We can check the label of this line...which is the zeroth column, that we previously excluded and took only pixels
M[0,0]
```

```python
# We can change the color of the image to gray, i.e black & white image
plt.imshow(im, cmap="gray")
plt.show()
```

```python
# We can also negative this black & white image...i.e we can change the blacks to white and vice versa
plt.imshow(255-im, cmap="gray")
plt.show()
```
