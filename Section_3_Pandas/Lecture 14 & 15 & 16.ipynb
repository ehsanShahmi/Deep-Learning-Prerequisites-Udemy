But what if we want different values in different rows of the dataframe? there comes apply() function

# The Apply() Function

The value of this new column is to be derived from the other columns.

suppose we want x1*x2 as new column...this needs: 

df["x1x2"] = df.apply(lambda row: row["x1"] * row["x2"], axis=1)

axis=1 means this applies to each ROW, rather than each COLUMN.

lambda is equivalent to :-->  def poop(row):return row["x1"] * row["x2"] ... df["x1x2"] = df.apply(poop, axis=1)


#### This apply() function actually does this:
interactions = []

for idx, row in df.iterrows() : 

    x1x2 = row["x1"] * row["x2"]
    interactions.append(x1x2)
df["x1x2"] = interactions

But this is very slow, so we never use this.

#### Now we want to create a new column with separated date & month argumnts derived from the first column:

```python
from datetime import datetime
```

```python
# Now to strip a time, just to test our function: 
datetime.strptime("1949-05","%Y-%m")
```

```python
# NOW, we use our apply() methods to newly create the new column:
df["dt"] = df.apply(lambda row: datetime.strptime(row["month"],"%Y-%m") , axis=1)
df.info()
```

Thus apply() method can be used to add newly column with derived different values.
