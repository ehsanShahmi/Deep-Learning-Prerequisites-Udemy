# Sampling from a general Multivariate Normal Distribution

We sample from a general Multivariate Normal Distribution...here, dimensions are not indepent of each other...for this, we need a full separate covariance matrix.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = "12, 8"
from scipy.stats import multivariate_normal as mvn
```

```python
# The covariance matrix:
cov = np.array([[1, 0.8],[0.8,3]])
# This implies--> Variance of 1 for 1st dimension, variance of 3 in 2nd dimension, cov betwwen them is 0.8
```

```python
# defining the mu for the covariance:
mu = np.array([0,2])
```

```python
# So now we create some values using rvs function with large number of random data-points
r = mvn.rvs(mean=mu, cov=cov, size=1000)
```

```python
#We scatterplot to confirm our correctness
plt.scatter(r[:,0], r[:,1])
```

```python
plt.axis("equal")
plt.show()
```

So, we see that there is a little bit of correlation. Note that we can also create this correlation using numpy.

```python
r = np.random.multivariate_normal(mean=mu, cov=cov, size=1000)
plt.scatter(r[:,0],r[:,1])
plt.axis('equal')
plt.show()
```

Thus everything remains the same also with using numpy.

```python
# Lets demonstatre these filtering using numpy using sine curves with different frequencies:
x = np.linspace(0, 100, 10000)
y = np.sin(x) + np.sin(3*x) + np.sin(5*x)
plt.plot(y)
plt.show()
```

```python
# We are also showing the Frequency Fourier transform (fft) for this y:
Y = np.fft.fft(y)
# Since this works with complex numbers, so we need to find the magnitudes b4 plotting the graph
plt.plot(np.abs(Y))
plt.show()
```
