# Sampling from a Gaussian 1D distribution

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = "12, 8"
```

We have discussed previously, how to sample from standard normal distribution:::--->>>

```python
r = np.random.randn(10000)
plt.hist(r, bins=100)
plt.show()
```

Thus this shows that we sampled from staandard normal distribution.....using more number of buckets, thus showing clearly.

Now, we want to sample from a Gaussian distribution with an arbitrary given mean and given variance.

```python
# We can do this with same function, just giving a scale and adding a mean
r = 10 * np.random.randn(10000) + 5   #stndardDeviation = 10 & mean=5
plt.hist(r, bins=100)
plt.show()
```

This expression of r is just reverse of converting from non-standard to standard.

Also this also gives us the required curve.

# Sampling from a Gaussian distribution (Spherical & Elliptical)

We use realistic dataset where there r more than one dimensions. Known as spherical database. same function but jst added dimension.

One coordinate is unrelated to the other. So thats why, they appear as a sphere.

```python
r = np.random.randn(10000, 2)
# We can verify that these are scattered  like a circlurar shape using scatterplot.
plt.scatter(r[:,0], r[:,1])
plt.show()
```

For Elliptical Gaussian, the variance is different for each dimension that exists. 

```python
# Suppose we want stnd.Deviation of 2nd dimension =5 and its mean =2
# Can be easily done by just scaling:
r[:,1] = 5 * r[:,1] + 2
# scatter plotting again to test
plt.scatter(r[:,0], r[:,1])
plt.show()
```

Here, still the figure looks same, bcoz matplotlib scales the axes.

```python
# So if we equalize the axes, then we can see the difference....
# since the axes wud not be scaled anymore for gud visualization
plt.axis("equal")
plt.scatter(r[:,0], r[:,1])
plt.show()
```

Thus we can see now that the elliptical distribution occurs for dimensions with different variances.
